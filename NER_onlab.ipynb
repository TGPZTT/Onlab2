```python
from langchain_core.runnables import RunnableLambda
from langchain_nvidia_ai_endpoints import ChatNVIDIA
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from typing import List, Union
from operator import itemgetter

## Zero-shot osztályozás
zsc_prompt = ChatPromptTemplate.from_messages([
    ("system", (
        "Output the most relevant topic option. Please answer in few or one word! [Options : {options}]"
    )),
    ("user", "{input} = ")
])

## számozott kimenetek kezelése
def EnumParser(*idxs):
    '''Method that pulls out values from a mistral model that outputs numbered entries'''
    idxs = idxs or [slice(0, None, 1)]
    entry_parser = lambda v: v if (' ' not in v) else v[v.index(' '):]
    out_lambda = lambda x: [entry_parser(v).strip() for v in x.split("\n")]
    return StrOutputParser() | RunnableLambda(lambda x: itemgetter(*idxs)(out_lambda(x)))

## instruct model betöltése
instruct_llm = ChatNVIDIA(model="mistral_7b") | EnumParser(0)

zsc_chain = zsc_prompt | instruct_llm

```

```python
def zsc_call(input, options=["financial", "technological", "historical", "corporational"]):
    return zsc_chain.invoke({"input" : input, "options" : options})

print("-" * 50)
print(zsc_call("Innovation is booming, with new ideas and processes constantly emerging to streamline daily tasks. "))

print("-" * 50)
print(zsc_call("Ancient ruins offer a glimpse into past civilizations, their structures whispering stories of forgotten eras."))

print("-" * 50)
print(zsc_call("A company's culture shapes its success, fostering collaboration and driving results."))

print("-" * 50)
print(zsc_call("Smart investments can secure a brighter future, allowing you to reach your long-term goals."))
```

```python
from langchain_nvidia_ai_endpoints import ChatNVIDIA, NVIDIAEmbeddings

from langchain_community.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import ArxivLoader

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000, chunk_overlap=100,
    separators=["\n\n", "\n", ".", ";", ",", " ", ""],
)


print("A dokumentumok letöltése")
docs = [
    ArxivLoader(query="2301.01277").load(),  ## Consumer acceptance of the use of artificial intelligence in online shopping: evidence from Hungary
    ArxivLoader(query="2111.14521").load(),  ## Do soda taxes affect the consumption and health of school-aged children? Evidence from France and Hungary
    ArxivLoader(query="1901.00283").load(),  ## Digital Economy And Society. A Cross Country Comparison Of Hungary And Ukraine
    ArxivLoader(query="1812.11488").load(),  ## E-commerce in Hungary: A Market Analysis
]

## A hivatkozások lerövidítése.

for doc in docs:
    content = doc[0].page_content
    if "References" in content:
        doc[0].page_content = content[:content.index("References")]

## A dokumentumok felosztása és a túl rövid darabok kiszűrése.
print("Dokumentumfelosztás")
docs_chunks = [text_splitter.split_documents(doc) for doc in docs]
docs_chunks = [[c for c in dchunks if len(c.page_content) > 200] for dchunks in docs_chunks]

## big-picture részletek miatt néhány nagyobb darabot is készítünk
doc_string = "Betöltött dokumentumok (angol nyelvű):"
doc_metadata = []
for chunks in docs_chunks:
    metadata = getattr(chunks[0], 'metadata', {})
    doc_string += "\n - " + metadata.get('Title')
    doc_metadata += [str(metadata)]

extra_chunks = [doc_string] + doc_metadata

## Printing out some summary information for reference
pprint(doc_string, '\n')
for i, chunks in enumerate(docs_chunks):
    print(f"Document {i}")
    print(f" - Metadata: {chunks[0].metadata}")
    print(f" - # Chunks: {len(chunks)}")
    print()
```

A dokumentumból készült vektorokat el kell tárolni.
Ezt a FIASS vektortároló végzi.
 https://python.langchain.com/docs/integrations/vectorstores/faiss

 Ez a FIASS tárolja a beágyazott modellt és a kéréseket tárolni, hogy egy kereshető indexet tudjon adni.

 Ennek kapcsán az input query-ben kapott dokumentumokra tud majd hivatkozni a LangChain API-n keresztül.

```python

from faiss import IndexFlatL2
from langchain_community.docstore.in_memory import InMemoryDocstore

from langchain_core.prompts import ChatPromptTemplate

embedder = NVIDIAEmbeddings(model="nvolveqa_40k", model_type=None) #Más model is lehet.

#Dokumentumtároló létrehozása - egyszerű összefűzéssel
print("Dokumentumtárolók létrehozása")
vecstores = [FAISS.from_texts(extra_chunks, embedder)]
vecstores += [FAISS.from_documents(doc_chunks, embedder) for doc_chunks in docs_chunks]
```

Az indexeket egyesíteni kell

```python
embed_dims = len(embedder.embed_query("test"))
def default_FAISS():
    return FAISS(
        embedding_function=embedder,
        index=IndexFlatL2(embed_dims),
        docstore=InMemoryDocstore(),
        index_to_docstore_id={},
        normalize_L2=False
    )

## Egy teljesen új (üres) FAISS Index létrehozása és ebbe gyűjtjük össze
def aggregate_vstores(vectorstores):
    agg_vstore = default_FAISS()
    for vstore in vectorstores:
        agg_vstore.merge_from(vstore) ## a merge_from optimalizál is egyben!!!
    return agg_vstore

if 'docstore' not in globals():
    docstore = aggregate_vstores(vecstores)

print(f"Elkészült az aggregált adatbázis {len(docstore.docstore._dict)} darabból")
```

```python
from langchain.document_transformers import LongContextReorder
```

```python
def RPrint(preface=""):
    """Simple passthrough "prints, then returns" chain"""
    def print_and_return(x, preface):
        print(f"{preface}{x}")
        return x
    return RunnableLambda(partial(print_and_return, preface=preface))

def docs2str(docs, title="Document"):
    """Useful utility for making chunks into context string. Optional, but useful"""
    out_str = ""
    for doc in docs:
        doc_name = getattr(doc, 'metadata', {}).get('Title', title)
        if doc_name:
            out_str += f"[Quote from {doc_name}] "
        out_str += getattr(doc, 'page_content', str(doc)) + "\n"
    return out_str

#A hosszabb dokumentumokat a kimeneti szöveg közepére rendezi át.
long_reorder = RunnableLambda(LongContextReorder().transform_documents)
```

```python
from langchain_core.runnables import RunnableLambda
from langchain_core.runnables.passthrough import RunnableAssign

from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

import gradio as gr
from functools import partial
from operator import itemgetter


llm = ChatNVIDIA(model="mixtral_8x7b") | StrOutputParser()
convstore = default_FAISS()

#Néhány előre elkészített parancs - prompt engineering
def save_memory_and_get_output(d, vstore):
    """Accepts 'input'/'output' dictionary and saves to convstore"""
    vstore.add_texts([
        f"User previously responded with {d.get('input')}",
        f"Agent previously responded with {d.get('output')}"
    ])
    return d.get('output')

initial_msg = (
    "Hello! I am a document chat agent here to help the user!"
    f" I have access to the following documents: {doc_string}\n\nHow can I help you?"
)

chat_prompt = ChatPromptTemplate.from_messages([("system",
    "You are a document chatbot. Help the user as they ask questions about documents."
    " User messaged just asked: {input}\n\n"
    " From this, we have retrieved the following potentially-useful info: "
    " Conversation History Retrieval:\n{history}\n\n"
    " Document Retrieval:\n{context}\n\n"
    " (Answer only from retrieval. Only cite sources that are used. Make your response conversational.)"
), ('user', '{input}')])

retrieval_chain = (
    {'input' : (lambda x: x)}

    | RunnableAssign({'history' : itemgetter('input') | convstore.as_retriever() | long_reorder | docs2str})
    | RunnableAssign({'context' : itemgetter('input') | docstore.as_retriever()  | long_reorder | docs2str})
    | RPrint()
)



stream_chain = chat_prompt | llm

def chat_gen(message, history=[], return_buffer=True):
    buffer = ""
    ## Az input üzenet alapján végezze el a kezelést
    retrieval = retrieval_chain.invoke(message)
    line_buffer = ""

    ## Ezután streameljük a stream_chain eredményeit.
    for token in stream_chain.stream(retrieval):
        buffer += token
        ## A sima print formázása - túl hosszú volna egy-egy sor. stringműveletek
        if not return_buffer:
            line_buffer += token
            if "\n" in line_buffer:
                line_buffer = ""
            if ((len(line_buffer)>84 and token and token[0] == " ") or len(line_buffer)>100):
                line_buffer = ""
                yield "\n"
                token = "  " + token.lstrip()
        yield buffer if return_buffer else token

    ## A chatet elmentjük a memóriába
    save_memory_and_get_output({'input':  message, 'output': buffer}, convstore)


```

```python
## Start kérdés
test_question = "Tell me about Hungary!"

## Before you launch your gradio interface, make sure your thing works
for response in chat_gen(test_question, return_buffer=False):
    print(response, end='')
```

```python
from transformers import TextDataset, DataCollatorForLanguageModeling
from transformers import GPT2Tokenizer, GPT2LMHeadModel
from transformers import Trainer, TrainingArguments
```

```python
def adathalmaz(utvonal, tokenizer, block_meret = 128):
    adathalmaz = TextDataset(
        tokenizer = tokenizer,
        file_path = utvonal,
        block_size = block_meret,
    )
    return adathalmaz


def adat_collator(tokenizer, mlm = False):
    adat_collator = DataCollatorForLanguageModeling(
        tokenizer=tokenizer,
        mlm=mlm,
    )
    return adat_collator


def train(model_name,
          trening_szoveg,
          output_dir,
          per_device_train_batch_size,
          num_train_epochs,
          save_steps):
  tokenizer = GPT2Tokenizer.from_pretrained(model_name)
  train_dataset = adathalmaz(trening_szoveg, tokenizer)
  data_collator = adat_collator(tokenizer)

  tokenizer.save_pretrained(output_dir)

  model = GPT2LMHeadModel.from_pretrained(model_name)

  model.save_pretrained(output_dir)

  training_args = TrainingArguments(
          output_dir=output_dir,
          overwrite_output_dir=False,
          per_device_train_batch_size=per_device_train_batch_size,
          num_train_epochs=num_train_epochs,
      )

  trainer = Trainer(
          model=model,
          args=training_args,
          data_collator=data_collator,
          train_dataset=train_dataset,
  )

  trainer.train()
  trainer.save_model()
```

```python
# Ez kb 10 perces művelet.
train(
      'NYTK/PULI-GPT-2',
      "/content/Articles_3.txt",
      '/content/ujmodel',
      10,
      10.0,
      450,
)
```

```python
from transformers import PreTrainedTokenizerFast, GPT2LMHeadModel, GPT2TokenizerFast, GPT2Tokenizer
```

```python
model_path = "/content/ujmodel"

def generate_text(input_text, max_length):

    model = GPT2LMHeadModel.from_pretrained(model_path)
    tokenizer = tokenizer = GPT2Tokenizer.from_pretrained(model_path)
    ids = tokenizer.encode(f'{input_text}', return_tensors='pt')
    output = model.generate(
        ids,
        do_sample=True,
        max_new_tokens=max_length,
        top_k=40,
        top_p=0.97,
        pad_token_id=model.config.eos_token_id,
    )
    return(tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
input_text = "A kedvenc szalámi márkám a Pick"
max_length = 40
riport = generate_text(input_text, max_length)
```

```python
def anonim(string):
    blacklist = ["Pick", "Nógrádi", "Herz", "Nike", "Stabilo"]
    cserelt_string = string
    for item in blacklist:
        cserelt_string = cserelt_string.replace(item, "*****")
    return cserelt_string

anon_string = anonim(riport)
print("Eredeti string:", riport)
print("Módosított string:", anon_string)
```
